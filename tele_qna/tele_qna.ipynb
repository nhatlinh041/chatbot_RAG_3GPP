```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("netop/TeleQnA")
```

```python
# save data set as json
import json

# Save the dataset to JSON files
for split_name, split_data in ds.items():
    output_file = f"tele_qna_{split_name}.json"
    split_data.to_json(output_file)
    print(f"Saved {split_name} split to {output_file}")

# Alternatively, save as a single JSON file with all splits
with open("tele_qna_full.json", "w") as f:
    json.dump({split: ds[split].to_dict() for split in ds.keys()}, f, indent=2)
print("Saved full dataset to tele_qna_full.json")
```
